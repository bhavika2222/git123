```python
import matplotlib.pyplot as plt
import numpy as np

# Replace this with your actual text data
text = "This is a sample text where you can find words like show, complete, welcome, famous, holiday, Champions, one, girls, parliament."

tokens = text.split()

frequency_types = {}
for token in tokens:
    if token in frequency_types:
        frequency_types[token] += 1
    else:
        frequency_types[token] = 1

words_chosen = ["show", "complete", "welcome", "famous", "holiday", "Champions", "one", "girls", "parliament"]
axes = []

for word in words_chosen:
    if word in frequency_types:
        axes.append([len(word), frequency_types[word], word])

axes.sort()

x = [item[0] for item in axes]
y = [item[1] for item in axes]

plt.plot(x, y, color='maroon')
plt.title("Zipf's law of length")
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```
